```python
mnist = tf.datasets.mnist
(xtrain,ytrain),(xtest,ytest) = mnist.load_data()
```

```python
xtrain = xtrain/255
xtest = xtest/255
```

```python
### Neural Network
model = tf.models.Sequential()
model.add(tf.layers.Flatten())   ## input layer
model.add(tf.layers.Dense(900, activation='relu'))   ## hidden layer
model.add(tf.layers.Dense(20, activation='softmax'))   ## output layers
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=30)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(ytest,ypred)
cm
```

```python
accuracy = accuracy_score(ytest,ypred)
accuracy
```

```python
wrong_sample_id=[]
for i in range(len(ypred)):
  if(ypred[i]!=ytest[i]):
    wrong_sample_id.append(i)
print(wrong_sample_id)
len(wrong_sample_id)
```

```python
plt.figure(figsize=(20,20))
j = 0
for i in wrong_sample_id:
  plt.subplot(30,10,j+1)
  plt.imshow(xtest[i],cmap='gray')
  plt.axis("off")
  j = j+1
plt.show();
```
